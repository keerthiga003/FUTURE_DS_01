```python
#Data cleaning
#1.check missing value

df.isnull().sum()
```

```python
df['Postal Code'] = df['Postal Code'].fillna(0)
```

```python
df = df.drop_duplicates()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.rename(columns={
    'Sales Amount': 'Sales'
}, inplace=True)
```

```python
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Month Name'] = df['Order Date'].dt.strftime('%B')
```

```python
#TOTAL SALES AND PROFIT

total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()

total_sales, total_profit
```

```python
#SALES BY YEAR

sales_by_year = df.groupby('Order Year')['Sales'].sum().reset_index()
sales_by_year
```

```python
plt.figure(figsize=(8,5))
plt.plot(sales_by_year['Order Year'], sales_by_year['Sales'], marker='o')
plt.title("Sales Trend by Year")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```

```python
#MONTHLY SALES TRENDS

monthly_sales = df.groupby('Month Name')['Sales'].sum().sort_values()
monthly_sales
```

```python
monthly_sales.plot(kind='bar', figsize=(10,5), title='Monthly Sales Trend')
plt.show()
```

```python
#CATEGORY-WISE SALES
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
category_sales

```

```python
category_sales.plot(kind='bar', figsize=(8,5), title='Sales by Category')
plt.show()
```

```python
#Sub-Category Performance

sub_category_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
sub_category_sales.head(10)
```

```python
sub_category_sales.plot(kind='bar', figsize=(8,5), title='Sales by Sub_Category')
plt.show()
```

```python
#Region-wise Sales

region_sales = df.groupby('Region')['Sales'].sum()
region_sales
```

```python
region_sales.plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title("Sales by Region")
plt.show()
```

```python
#Top 10 Products by Sales

top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_products
```

```python
top_products.plot(kind='barh', figsize=(8,6))
plt.title("Top 10 Products by Sales")
plt.show()
```

```python
#Profit Analysis

profit_by_category = df.groupby('Category')['Profit'].sum()
profit_by_category
```

```python
profit_by_category.plot(kind='bar', title='Profit by Category')
plt.show()
```
