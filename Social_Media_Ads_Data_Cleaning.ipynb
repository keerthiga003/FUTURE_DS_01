```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Mariselvam\Downloads\archive (1)\marketing_data.csv")
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.columns
```

```python
df.isnull().sum()
```

```python
df['income'] = (
    df['income']
    .astype(str)
    .str.replace(',', '', regex=True)
    .str.replace('$', '', regex=True)
    .str.replace(' ', '', regex=True)
    .replace(['NA', 'N/A', '', 'unknown'], np.nan)
)
```

```python
df['income'] = pd.to_numeric(df['income'], errors='coerce')
```

```python
df['income'].describe()
```

```python
df['income'].fillna(0, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['dt_customer'].head(10)
```

```python
df['dt_customer'] = pd.to_datetime(
    df['dt_customer'],
    format='%Y-%m-%d',
    errors='coerce'
)
```

```python
df['dt_customer'].isna().sum()
```

```python
df['age'] = 2025 - df['year_birth']
```

```python
df['total_spend'] = (
    df['mntwines'] +
    df['mntfruits'] +
    df['mntmeatproducts'] +
    df['mntfishproducts'] +
    df['mntsweetproducts'] +
    df['mntgoldprods']
)
```

```python
df['total_purchases'] = (
    df['numwebpurchases'] +
    df['numcatalogpurchases'] +
    df['numstorepurchases']
)
```

```python
campaign_acceptance_rate = (df['response'].sum() / len(df)) * 100
campaign_acceptance_rate
```

```python
campaign_cols = [
    'acceptedcmp1', 'acceptedcmp2', 'acceptedcmp3',
    'acceptedcmp4', 'acceptedcmp5'
]

df[campaign_cols].sum()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['total_spend'], bins=30)
plt.title('Total Spend Distribution')
plt.show()
```

```python
product_spend = df[['mntwines','mntfruits','mntmeatproducts',
                    'mntfishproducts','mntsweetproducts','mntgoldprods']].sum()

product_spend.plot(kind='bar', figsize=(8,5))
plt.title('Revenue by Product Category')
plt.show()
```

```python
sns.countplot(x='response', data=df)
plt.title('Campaign Response Distribution')
plt.show()
```
